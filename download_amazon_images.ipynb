```python
import time

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36'}

def get_links(url_format, out_file, max_pages=400):
    with open(out_file, 'w') as f:
        for i in range(max_pages):
            url = url_format.format(i)
            
            try:
                r = requests.get(url, headers=headers)
            except:
                print(url)
                continue
            
            soup = bs(r.text)
            imgs = soup('img', class_='s-access-image')
            print(i, len(imgs))
            for img in imgs:
                f.write(img['src']+'\n')
            time.sleep(1)
                
get_links('https://www.amazon.com/s/?rh=n%3A5174%2Cn%3A%21301668%2Cn%3A67207&page={}', out_file='links/metal_amazon.csv')
```

```python
url_formats = {'country': 'https://www.amazon.com/s/?rh=n%3A5174%2Cn%3A%21301668%2Cn%3A16&page={}',
               'classical': 'https://www.amazon.com/s/?rh=n%3A5174%2Cn%3A%21301668%2Cn%3A85&page={}',
               'rock': 'https://www.amazon.com/s/?rh=n%3A5174%2Cn%3A%21301668%2Cn%3A40&page={}',
               'jazz': 'https://www.amazon.com/s/?rh=n%3A5174%2Cn%3A%21301668%2Cn%3A34&page={}',
               'rap': 'https://www.amazon.com/s/?rh=n%3A5174%2Cn%3A%21301668%2Cn%3A38&page={}',
               'pop': 'https://www.amazon.com/s/?rh=n%3A5174%2Cn%3A%21301668%2Cn%3A37&page={}'}

for k, v in url_formats.items():
    filename = 'links/{}_amazon.csv'.format(k)
    print(filename)
    get_links(v, filename, max_pages=100)
```

```python
import h5py
from PIL import Image
from io import BytesIO
import numpy as np
from matplotlib.pyplot import imshow
%matplotlib inline


def fetch_images(genre_name):
    link_filename = 'links/{}_amazon.csv'.format(genre_name)
    with open(link_filename, 'r') as f:
        links = f.read().split()
        links = list(set(links[2:])) # remove dups, remove top images as well

    dataset = []
    labels = []
    for link in links:
        r = requests.get(link)
        img = Image.open(BytesIO(r.content))
        img = img.resize((150,150))
        
        array = np.asarray(img) / 255
        if array.shape == (150, 150, 3):
            dataset.append(array)
        
            if genre_name == 'metal':
                labels.append(1.)
            else:
                labels.append(0.)
    
    dataset = np.stack(dataset)
    dataset = dataset.astype('float32')
    
    labels = np.asarray(labels)
    labels = labels.astype('float32')
    
    return dataset, labels

X, y = fetch_images('metal')
print(X.shape, y.shape)
with h5py.File('larger_dataset.hdf5', 'a') as f:
    f.create_dataset('metal_X', data=X)
    f.create_dataset('metal_y', data=y)
```

```python
for genre in url_formats.keys():
    print('starting:', genre)
    X, y = fetch_images(genre)
    print(X.shape, y.shape)
    with h5py.File('larger_dataset.hdf5', 'a') as f:
        f.create_dataset('{}_X'.format(genre), data=X)
        f.create_dataset('{}_y'.format(genre), data=y)
    print('finished:', genre)
```
